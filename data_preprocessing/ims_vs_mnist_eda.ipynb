```python
import pandas as pd
import numpy as np

def calculate_variances(df):
    # Extract labels and data from the DataFrame
    labels = df['Label'].values
    data = df.drop(columns=['Label']).values
    
    classes = np.unique(labels)
    n = data.shape[0]
    overall_mean = np.mean(data, axis=0)
    
    # Within-Class Variance
    within_class_variance = 0
    for c in classes:
        class_data = data[labels == c]
        class_mean = np.mean(class_data, axis=0, keepdims=True)
        n_c = class_data.shape[0]
        sample_variance = np.sum(np.square(class_data - class_mean), axis=1)
        within_class_variance = np.sum(sample_variance)/(n_c-1)
        var = class_data.var()
        # within_class_variance = np.sum(feature_variance)/class_data.shape[1]
        # within_class_variance = np.sum(np.square(class_data - class_mean), axis=1)/(n_c)
        # within_class_variance = np.sum((class_data - class_mean)**2)/(n_c-1)
        # within_class_variance = np.sum(np.square(class_data - class_mean))/(n_c)
        # within_class_variance = np.sum(class_data.subtract(series, axis=0)/(n_c-1))
        var = class_data.var()
        print(within_class_variance)
        print(var)
        var_ = np.var(class_data)
        print(var_)
        print('-----------------')
    
    # Between-Class Variance
    between_class_variance = 0
    # for c in classes:
    #     class_data = data[labels == c]
    #     class_mean = np.mean(class_data, axis=0)
    #     n_c = class_data.shape[0]
    #     # print(n_c, 1/n_c)
    #     between_class_variance += np.sum((class_mean - overall_mean) ** 2)/(1/(n_c-1)) 
    #     # print(round((1/n_c) * np.sum((class_mean - overall_mean) ** 2)), round(n_c * np.sum((class_mean - overall_mean) ** 2)))
    
    return within_class_variance, between_class_variance
```
