```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../data/ion_mobility_paper/ion_mobility_paper_data_w_inchikeys_and_class_info.csv')
embeddings = pd.read_csv('../data/ion_mobility_paper/ion_mobility_paper_embeddings.csv')
```

```python
data.head()
```

```python
fig, ax = plt.subplots(figsize=(14, 8))
slice_num = 100

sns.scatterplot(data=data[:][:slice_num], x='m/z', y='Class', hue='Class', legend="brief", ax=ax, s=100)
sns.move_legend(ax, 'upper right')

# plt.xticks([])
plt.xlabel('Mass to Charge Ratio (m/z)', fontsize=24)
# plt.ylabel('Compound Name', fontsize=16)
plt.ylabel('Compound Name', fontsize=24)
plt.title('CCS Average Values Across Molecules', fontsize=28)
plt.grid()
plt.show()
```
